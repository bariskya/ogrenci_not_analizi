```python
df["any pass?"]=df.apply(lambda x: "F" if x["math"]== "F" or x["reading"]== "F" or x["writing"]== "F" else "P" ,axis=1)
df["any pass?"].value_counts() # ÖĞRENCİ HERHANGİ BİRİNDEN KALMIŞMI
```

```python
df["percentage"]=(df["math score"]+df["reading score"]+df["writing score"]/3).round()
df["total marks"]=df["math score"]+df["reading score"]+df["writing score"]
```

```python
def grades(percentage,anypass== "F"):
       return "F"
    if (percentage >= 80):
         return "A"
    if (percentage >= 70):
         return "B"
    if (percentage >= 60):
         return "C"
    if (percentage >= 50):
         return "D"
    if (percentage >= 40):
         return "E"

    else:
         return "F"

# DataFrame'e grades sütununu ekleyelim
df["grades"] = df["percentage"].apply(grades)

# Notların sayısını gösterelim
print(df["grades"].value_counts())
```

```python
male=df["gender"].value_counts()["male"]
female=df["gender"].value_counts()["female"]
male_ratio=male/(female+male)*100
female_ratio=female/(female+male)*100
print("female raito : ",female_ratio)
print("male raito : ",male_ratio)                      #KADIN ERKEK ORANLARI
```

```python
# Pie grafiği
labels = ['Female', 'Male']
sizes = [female_ratio, male_ratio]

# Grafiği çiz
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Gender Ratio')
plt.axis('equal')  # Grafik yuvarlak olsun
plt.show()
```

```python
etnik=df.groupby(["race/ethnicity","parental level of education"])["math score"].mean().reset_index()
sns.catplot(x="parental level of education",y="math score",hue="race/ethnicity",kind="bar",data=etnik)
plt.title("etnik kökene göre mat skoru")
plt.xlabel("etnik")                                         # ETNİK KÖKENLERE GÖRE MAT NOTU
plt.ylabel("math score")
plt.xticks(rotation=90)

plt.show()
```
